```python
import requests
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd

import re
import time
```

```python
url="https://www.flipkart.com/search?q=earbuds%20above%201000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
page=requests.get(url)
soup = BeautifulSoup(page.text)


p=soup.find_all("div",class_="yRaY8j")
Originalprice = []
for i in p:
    Originalprice.append(i.text)


p1=soup.find_all("div",class_="Nx9bqj")
Discountedprice=[]
for i in p1:
    Discountedprice.append(i.text)


data=soup.find_all("a",class_="wjcEIp")
Brands = []
for i in data:
    Brands.append(i.text.split()[0])

Models=[]
for i in data:
    w=i.text.split()[0:3] 
    model= " ".join(w)
    Models.append(model)


d1=soup.find_all("div",class_="XQDdHH")
Rating = []
for i in d1:
    Rating.append(i.text)


Battery = []
for i in data:
    text =i.text
    match = re.findall(r'(\d+)\s*(?:[hH](?:rs?|ours?)?|hr|hours?|hrs?)\s*(?:[Bb]attery|[Pp]laytime|[Pp]layback)?', text)
    if match:
        Battery.append(match[0])
    else:
        Battery.append(np.nan)

Discount = []
for tag in soup.find_all("div", class_="UkUFwK"):
    text = tag.text.replace('off', '').strip()
    Discount.append(text)



d4=soup.find_all("span",class_="Wphh3N")
Reviews = []
for i in d4:
    Reviews.append(i.text)
    
```

```python
d = {
    "Brands": Brands,
    "Models": Models,
    "Battery": Battery,
    "OriginalPrice": Originalprice,
    "DiscountedPrice": Discountedprice,
    "Discount": Discount,
    "Rating": Rating,
    "Reviews": Reviews
}
```

```python
cols = [Brands, Models, Battery, Originalprice, Discountedprice, Discount, Rating, Reviews]
max_len = max(len(lst) for lst in cols)

for lst in cols:
    while len(lst) < max_len:
        lst.append(np.nan)
```
