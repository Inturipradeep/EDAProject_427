```python
data["Battery"] = data["Battery"].astype(int)
data["OriginalPrice"] = pd.to_numeric(data["OriginalPrice"].astype(str).str.replace("₹","").str.replace(",",""), errors='coerce').fillna(0).astype(int)
data["Discount"]=data["Discount"].astype(str).str.replace("%","").str.replace(",","").astype(int)
data["DiscountedPrice"] = data["DiscountedPrice"].astype(str).str.replace("₹","").str.replace(",","").replace("nan", np.nan).fillna(0).astype(float)
data["Reviews"]=data["Reviews"].fillna(0).astype(str).str.replace("(", "").str.replace(")", "").str.replace(",", "").fillna(0).astype(int)
```

```python
### Comparing how much discount each brand usually gives.
data['DiscountPercent'] = ((data['OriginalPrice'] - data['DiscountedPrice']) / data['OriginalPrice']) * 100

plt.figure(figsize=(10,6))
sns.barplot(data=data, x='Brands', y='DiscountPercent', palette='coolwarm', estimator='mean')
plt.title('Average Discount % by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Discount (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
## Comparing customer satisfaction across brands.
plt.figure(figsize=(10,6))
sns.barplot(data=data, x='Brands', y='Rating', estimator='mean', palette='Set2')
plt.title('Average Rating by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
from ipywidgets import interact
```

```python
@interact
def func(col1=data.select_dtypes("int64").columns,   
         col2=data.select_dtypes("object").columns):   
    return data.groupby([col2])[col1].agg(("min","max","median","mean"))
```

```python
@interact
def func(col1=data.select_dtypes("int64").columns, 
         col2=data.select_dtypes("object").columns):  
    sns.violinplot(y=data[col1], x=data[col2],palette=["green","blue"])
```
